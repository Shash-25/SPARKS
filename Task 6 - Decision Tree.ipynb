In this task, we will train a Decision Tree Classifier to predict species using features like SepalLength, SepalWidth, PetalLength and PetalWidth. 

### Importing Libraries

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
print()
```

### Importing the Data Set

```python
data = pd.read_csv("Iris.csv")

#storing dataframe into numpy arrays to use them in training the model
X = np.asanyarray(data[["SepalLengthCm", "SepalWidthCm","PetalLengthCm","PetalWidthCm"]])
Y = np.asanyarray(data[["Species"]])
data.head()
```

### Train-Test Splitting

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size= 0.2,random_state = 1)
print()
```

### Training the Model

```python
tree = DecisionTreeClassifier(criterion = "entropy", max_depth = 6)
tree.fit(X_train,Y_train)
#Now our model is trained!
```

### Accuracy

```python
y_predict = tree.predict(X_test)
accuracy_score(y_predict,Y_test)
```

### Visualizing the graph

```python
from six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
dot_data = StringIO()
export_graphviz(tree, out_file=dot_data, feature_names=data.columns.values[1:5],  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
